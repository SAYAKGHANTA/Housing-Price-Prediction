```python
# Missing Values
```

```python
#We will check the percentage of nan values in the features

#setp 1: make a list which include all the features which has nan values

nan_features_list=[]
for features in dataset.columns:
    if(dataset[features].isnull().sum()>1):
        nan_features_list.append(features)

# print(nan_features_list)

#step 2: check the percentage of missing values

for features in nan_features_list:
    print(features,"=", np.round(dataset[features].isnull().mean(),4), " % missing values")
```

```python
# Find relationship between missing values and sale-price
```

```python
# Plot relationship between features and sale-price
```

```python
for feature in nan_features_list:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.ylabel('sale price')
    plt.show()
```

```python
# we can see the relationship between missing values and dependent variable...we have to replace these nan values with something meaningful
```

```python

# Numeric values
```

```python
#numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

numerical_features=[]

for feature in dataset.columns:
    if(dataset[feature].dtypes!='O'):
        numerical_features.append(feature)


numerical_features
```

```python
dataset[numerical_features].head()
```

```python
dataset[numerical_features].describe()
```

```python
# Handle the temporary variable : like date-time vriable
```

```python
year_feature=[]

for feature in numerical_features:
    if("Yr" in feature or "Year" in feature):
        year_feature.append(feature)


year_feature
```

```python
# Now we will analyze the temporal date-time variable
# we will check if there is a relation between these features and saleprice

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('Median Sale-Price')
plt.title('House price vs. Year sold')
plt.show()
```

```python
# we will check the discrete features and continuous fatures in numerical features

# 1. Discrete features

discrete_features=[feature for feature in numerical_features if len(dataset[feature].unique())<10 and feature not in year_feature+['Id']]
discrete_features
len(discrete_features)
```

```python
dataset[discrete_features].head()
```

```python
# Lets Find the realtionship between discrete features and Sale-price

for feature in discrete_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
# Continuous Features
continuous_features=[feature for feature in numerical_features if feature not in discrete_features +year_feature+['Id']]
continuous_features
```

```python
# Lets Find the realtionship between continuous features and Sale-price

for feature in continuous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
#Lets use logarithmic transformation


for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```
